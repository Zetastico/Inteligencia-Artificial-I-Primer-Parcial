```python
import pandas as pd
import numpy as np

# Cargar el archivo XPT
path = r'D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D3/DEMO_L.xpt'
df = pd.read_sas(path)

# Lista de columnas a eliminar (Identificadores y pesos estadísticos)
# SEQN: ID único (no aporta al aprendizaje)
cols_to_drop = ['SEQN', 'WTINT2YR', 'WTMEC2YR', 'SDMVSTRA', 'SDMVPSU']
df_ml = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

# Limpieza de nulos
# Como queremos usar todas las columnas, aplicaremos una estrategia automática:
for col in df_ml.columns:
    # Si la columna tiene más del 70% de nulos, mejor la quitamos 
    if df_ml[col].isnull().sum() > (len(df_ml) * 0.7):
        df_ml = df_ml.drop(columns=[col])
    else:
        # Para el resto, usamos la mediana
        df_ml[col] = df_ml[col].fillna(df_ml[col].median())

df_final = df_ml.astype(float)

df_final.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D3/Nhanes_full_prepared.csv', index=False)

print(f"Dataset NHANES preparado.")
print(f"Columnas originales: {df.shape[1]}")
print(f"Columnas conservadas para el modelo: {df_final.shape[1]}")
print(f"Total de registros: {len(df_final)}")
```

```python
import pandas as pd
import numpy as np

# Cargar el archivo XPT original
path = r'D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D3/DEMO_L.xpt'
df_full = pd.read_sas(path)

# Limpieza de Nulos (para todas las columnas)
# Recorremos cada columna para asegurar que no queden casillas vacías
for col in df_full.columns:
    # Si la columna tiene nulos, los llenamos con la mediana
    if df_full[col].isnull().any():
        # Si la columna está 100% vacía, le ponemos 0
        if df_full[col].isnull().all():
            df_full[col] = df_full[col].fillna(0)
        else:
            df_full[col] = df_full[col].fillna(df_full[col].median())

# 3. Asegurar formato numérico
df_full = df_full.astype(float)

# 4. Guardar la versión completa
df_full.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D3/Nhanes_full_master.csv', index=False)

print("¡Dataset Maestro de NHANES generado!")
print(f"Total de columnas procesadas: {df_full.shape[1]}")
print(f"Total de filas: {len(df_full)}")
```
