```python
import pandas as pd
import numpy as np

df = pd.read_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D1/train.csv')

numeric_cols = df.select_dtypes(include=[np.number]).columns
categorical_cols = df.select_dtypes(exclude=[np.number]).columns

# Limpieza de valores nulos
# Para los números usamos la mediana
for col in numeric_cols:
    # Si la columna está totalmente vacía la llenamos con 0
    if df[col].isnull().all():
        df[col] = df[col].fillna(0)
    else:
        df[col] = df[col].fillna(df[col].median())

# Para el texto (categóricas): marcamos como 'None'
for col in categorical_cols:
    df[col] = df[col].fillna('None')

cols_to_drop = ['Order', 'PID', 'Id']
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

# (One-Hot Encoding)
# get_dummies transforma el texto en columnas de 1s y 0s
df_numeric = pd.get_dummies(df, drop_first=True)

# Convertir todo a float
df_numeric = df_numeric.astype(float)

df_numeric.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D1/ames_processed.csv', index=False)

print("Funciona")
print(f"Columnas originales: {df.shape[1]}")
print(f"Columnas finales (tras conversión): {df_numeric.shape[1]}")
```
