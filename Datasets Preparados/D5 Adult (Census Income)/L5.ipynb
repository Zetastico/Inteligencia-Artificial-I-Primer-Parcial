```python
import pandas as pd
import numpy as np

column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 
    'marital-status', 'occupation', 'relationship', 'race', 'sex', 
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'
]

# Carga de datos
# skipinitialspace=True elimina los espacios después de las comas
# na_values='?' convierte los signos de interrogación en nulos reales (NaN)
df = pd.read_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D5/adult/adult.data', names=column_names, skipinitialspace=True, na_values='?')

# Limpieza de nulos: Moda para texto, Mediana para números
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col] = df[col].fillna(df[col].mode()[0])
        else:
            df[col] = df[col].fillna(df[col].median())

# Binarizar el objetivo
# Transformamos '<=50K' en 0 y '>50K' en 1
df['income'] = df['income'].apply(lambda x: 1 if '>50K' in str(x) else 0)

# Convertimos todo el texto a números mediante One-Hot Encoding
df_full = pd.get_dummies(df, drop_first=True)
df_full.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D5/adult/adult_full_prepared.csv', index=False)

# 1. 'fnlwgt': Es un control estadístico del censo, no tiene relación con el perfil de la persona.
# 2. 'education': Es redundante porque ya existe 'education-num' que es la misma escala en número.
cols_to_drop = ['fnlwgt', 'education', 'native-country']
df_optimized = df.drop(columns=cols_to_drop)

# Convertimos a números la versión optimizada
df_optimized_numeric = pd.get_dummies(df_optimized, drop_first=True)
df_optimized_numeric.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D5/adult/adult_optimized_model.csv', index=False)

print("¡Archivos generados con éxito!")
print(f"Columnas en versión Full: {df_full.shape[1]}")
print(f"Columnas en versión Optimizada: {df_optimized_numeric.shape[1]}")
```
