```python
import pandas as pd
import numpy as np

# Definir nombres de columnas (A1 a A15 + Clase)
column_names = [f'A{i}' for i in range(1, 16)] + ['class']

# Cargar datos

# na_values='?' convierte los signos de interrogación en nulos (NaN)
df = pd.read_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D6/credit+approval/crx.data', names=column_names, na_values='?')


# Identificamos columnas numéricas y categóricas
numeric_cols = df.select_dtypes(include=[np.number]).columns
categorical_cols = df.select_dtypes(exclude=[np.number]).columns

# Imputación para números (Mediana)
for col in numeric_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

# Imputación para texto (Moda)
for col in categorical_cols:
    if df[col].isnull().any():
        fill_val = df[col].mode()[0] if not df[col].mode().empty else 'Unknown'
        df[col] = df[col].fillna(fill_val)

# Transformar la clase objetivo ('+' es aprobado, '-' es rechazado) a 1 y 0
df['class'] = df['class'].map({'+': 1, '-': 0})

# Aplicamos One-Hot Encoding a todas las variables categóricas
df_full_numeric = pd.get_dummies(df, drop_first=True)
df_full_numeric = df_full_numeric.astype(float)
df_full_numeric.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D6/credit_full_prepared.csv', index=False)


# En este dataset, las columnas A2, A3, A8, A11, A14 y A15 son numéricas.
# La columna A14 (ZipCode/Código Postal) es famosa en este dataset por ser 
# irrelevante para la solvencia crediticia y añadir ruido.
cols_to_drop = ['A14'] 
df_optimized = df.drop(columns=cols_to_drop)

df_opt_numeric = pd.get_dummies(df_optimized, drop_first=True)
df_opt_numeric = df_opt_numeric.astype(float)
df_opt_numeric.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D6/credit_optimized_model.csv', index=False)

print("¡Archivos de Crédito generados!")
print(f"Columnas Full: {df_full_numeric.shape[1]}")
print(f"Columnas Optimizada: {df_opt_numeric.shape[1]}")
```
