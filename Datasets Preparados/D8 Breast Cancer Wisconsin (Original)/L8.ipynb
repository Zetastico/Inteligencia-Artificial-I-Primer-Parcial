```python
import pandas as pd
import numpy as np

# Definir nombres de columnas según wdbc.names
# ID, Diagnóstico, y luego 30 características (10 medidas base x 3 tipos: media, error, peor)
features = ['radius', 'texture', 'perimeter', 'area', 'smoothness', 
            'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension']
types = ['mean', 'se', 'worst']
column_names = ['id', 'diagnosis'] + [f"{f}_{t}" for t in types for f in features]
# 2. Cargar datos
df = pd.read_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D8/breast+cancer+wisconsin+original/wdbc.data', names=column_names)

# Binarizar el objetivo (Diagnosis)
# M (Maligno) = 1, B (Benigno) = 0
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Limpieza de nulos (aunque este dataset suele venir muy limpio)
df = df.fillna(df.median())

df_full = df.astype(float)
df_full.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D8/breast_cancer_full_prepared.csv', index=False)


# 1. 'id': El número de identificación del paciente no tiene relación con el cáncer.
# 2. 'worst' features: A veces se eliminan las columnas 'worst' o 'se' para evitar 
#    el sobreajuste, pero aquí borraremos solo el ID para mantener la potencia del diagnóstico.
#    También borraremos las columnas 'fractal_dimension' si buscamos simplicidad extrema, 
#    pero lo estándar es solo quitar el ID.
cols_to_drop = ['id']
df_optimized = df.drop(columns=cols_to_drop)

df_opt_numeric = df_optimized.astype(float)
df_opt_numeric.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D8/breast_cancer_optimized_model.csv', index=False)

print("¡Archivos de Cáncer de Mama generados!")
print(f"Columnas Full: {df_full.shape[1]}")
print(f"Columnas Optimizada: {df_opt_numeric.shape[1]}")
```
