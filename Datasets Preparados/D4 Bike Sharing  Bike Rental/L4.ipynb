```python
import pandas as pd
import numpy as np

# Cargar el dataset
df_bike = pd.read_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D4/train.csv')

# Ingeniería de variables (Extraer lo útil de la fecha)
df_bike['datetime'] = pd.to_datetime(df_bike['datetime'])
df_bike['hour'] = df_bike['datetime'].dt.hour
df_bike['month'] = df_bike['datetime'].dt.month
df_bike['day_of_week'] = df_bike['datetime'].dt.dayofweek # 0=Lunes, 6=Domingo
df_bike['year'] = df_bike['datetime'].dt.year

# ELIMINACIÓN DE COLUMNAS IMPOSIBLES O REDUNDANTES
# 'casual' y 'registered' -> Fuga de datos (Leakage)
# 'atemp' -> Redundante con 'temp'
# 'datetime' -> Ya no es necesaria tras la extracción
cols_to_drop = ['datetime', 'casual', 'registered', 'atemp']
df_clean = df_bike.drop(columns=cols_to_drop)

# Manejo de nulos
df_clean = df_clean.fillna(df_clean.median())


df_final = df_clean.astype(float)

df_final.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D4/bike_sharing_clean_model.csv', index=False)

print("¡Dataset optimizado para modelos de IA!")
print(f"Columnas resultantes: {df_final.columns.tolist()}")
```
