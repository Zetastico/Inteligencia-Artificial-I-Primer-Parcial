```python
import pandas as pd
import numpy as np

# Definir nombres de columnas (A1 a A14 + Clase)
column_names = [f'A{i}' for i in range(1, 15)] + ['class']
# 2. Cargar datos
# sep='\s+' detecta uno o más espacios como separador
df = pd.read_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D7/statlog+australian+credit+approval/australian.dat', names=column_names, sep='\s+')

# En este dataset los nulos son raros, pero aplicamos la misma lógica
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

# Como ya son números, solo nos aseguramos de que sean float para el modelo
df_full_numeric = df.astype(float)
df_full_numeric.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D7/australian_full_prepared.csv', index=False)

# Según la documentación (australian.doc), la columna A13 es el ZipCode (Código Postal)
# Al igual que en el dataset anterior, el código postal suele ser ruido para el crédito.
# También A1 (Género anonimizado) se suele quitar en modelos éticos para evitar sesgos.
cols_to_drop = ['A13', 'A1'] 
df_optimized = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

df_opt_numeric = df_optimized.astype(float)
df_opt_numeric.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D7/australian_optimized_model.csv', index=False)

print("¡Archivos de Crédito Australiano generados!")
print(f"Columnas Full: {df_full_numeric.shape[1]}")
print(f"Columnas Optimizada: {df_opt_numeric.shape[1]}")
```
