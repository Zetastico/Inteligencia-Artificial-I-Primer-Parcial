```python
import pandas as pd
import numpy as np

# Cargar el dataset
df = pd.read_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D9/mening missing 12.csv')

# Separamos por tipo de contenido real
numeric_cols = df.select_dtypes(include=[np.number]).columns
categorical_cols = df.select_dtypes(exclude=[np.number]).columns

# A. Imputación para números (Mediana)
for col in numeric_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

# B. Imputación para texto (Moda)
for col in categorical_cols:
    if df[col].isnull().any():
        fill_val = df[col].mode()[0] if not df[col].mode().empty else 'Unknown'
        df[col] = df[col].fillna(fill_val)

# Convertimos Risk_Level a una escala numérica (Ordinal Encoding)
# ya que el riesgo tiene un orden lógico: Low < Moderate < High
risk_map = {'Low Risk': 0, 'Moderate Risk': 1, 'High Risk': 2}
df['Risk_Level'] = df['Risk_Level'].map(risk_map)


df_full_numeric = pd.get_dummies(df, drop_first=True)
df_full_numeric = df_full_numeric.astype(float)
df_full_numeric.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D9/medical_risk_full.csv', index=False)


# 1. 'Patient_ID': Es un identificador único, no tiene valor predictivo (ruido puro).
# 2. 'Outcome': En un modelo de predicción de riesgo médico, el "desenlace" (si murió o se recuperó)
#    ocurre DESPUÉS del diagnóstico. Si lo dejas, el modelo hará trampa (Data Leakage).
# 3. 'Diagnosis': Si ya estamos prediciendo el riesgo basado en laboratorio, 
#    a veces el diagnóstico clínico es redundante, pero aquí quitaremos solo ID y Outcome.
cols_to_drop = ['Patient_ID', 'Outcome']
df_optimized = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

df_opt_numeric = pd.get_dummies(df_optimized, drop_first=True)
df_opt_numeric = df_opt_numeric.astype(float)
df_opt_numeric.to_csv('D:/Universidad/5to Semestre/Inteligencia Artificial/Primer Parcial/Datasets/D9/medical_risk_optimized.csv', index=False)

print("¡Datasets médicos generados con éxito!")
print(f"Columnas Full: {df_full_numeric.shape[1]}")
print(f"Columnas Optimizada: {df_opt_numeric.shape[1]}")
```
